```python
import numpy as np

from keras.layers import *
from keras.models import *



input_dimensi = len(X[1])

output_dimensi1 = len(one_hot_labelsYBarung1[1])

output_dimensi5 = len(one_hot_labelsYBarung5[1])

output_dimensi9 = len(one_hot_labelsYBarung9[1])

output_dimensi13 = len(one_hot_labelsYBarung13[1])

output_dimensi17 = len(one_hot_labelsYBarung17[1])

output_dimensi21 = len(one_hot_labelsYBarung21[1])

output_dimensi25= len(one_hot_labelsYBarung25[1])

output_dimensi29 = len(one_hot_labelsYBarung29[1])




inputs = Input(shape=(None,input_dimensi))

chanDim=-1
		# CONV => RELU => POOL
_ = Conv1D(filters=32, kernel_size=2, padding='same')(inputs)
_ = Activation("relu")(_)
#_ = BatchNormalization(axis=chanDim)(_)
_ = MaxPooling1D(pool_size= 1)(_)
#_ = Dropout(0.25)(_)
bottleneck = GlobalMaxPool1D()(_)
#bottleneck = Flatten()(_)

_ = Dense(units=128, activation='sigmoid')(bottleneck)
#_ = BatchNormalization()(_)
#_ = Dropout(0.25)(_)
output1 = Dense(output_dimensi1, activation='sigmoid', name='bonangbarung1')(_)


_ = Dense(units=128, activation='sigmoid')(bottleneck)
#_ = BatchNormalization()(_)
#_ = Dropout(0.25)(_)
output5 = Dense(output_dimensi5, activation='sigmoid', name='bonangbarung5')(_)


_ = Dense(units=128, activation='sigmoid')(bottleneck)
#_ = BatchNormalization()(_)
#_ = Dropout(0.25)(_)
output9 = Dense(output_dimensi9, activation='sigmoid', name='bonangbarung9')(_)


_ = Dense(units=128, activation='sigmoid')(bottleneck)
#_ = BatchNormalization()(_)
#_ = Dropout(0.25)(_)
output13 = Dense(output_dimensi13, activation='sigmoid', name='bonangbarung13')(_)


_ = Dense(units=128, activation='sigmoid')(bottleneck)
#_ = BatchNormalization()(_)
#_ = Dropout(0.25)(_)
output17 = Dense(output_dimensi17, activation='sigmoid', name='bonangbarung17')(_)



_ = Dense(units=128, activation='sigmoid')(bottleneck)
#_ = BatchNormalization()(_)
#_ = Dropout(0.25)(_)
output21 = Dense(output_dimensi21, activation='sigmoid', name='bonangbarung21')(_)



_ = Dense(units=128, activation='sigmoid')(bottleneck)
#_ = BatchNormalization()(_)
#_ = Dropout(0.25)(_)
output25 = Dense(output_dimensi25, activation='sigmoid', name='bonangbarung25')(_)



_ = Dense(units=128, activation='sigmoid')(bottleneck)
#_ = BatchNormalization()(_)
#_ = Dropout(0.25)(_)
output29 = Dense(output_dimensi29, activation='sigmoid', name='bonangbarung29')(_)










model = Model(inputs=inputs, outputs=[output1, output5, output9, output13, output17, output21, output25, output29])

model.compile(optimizer='adam', 
              loss={'bonangbarung1': 'binary_crossentropy','bonangbarung5': 'binary_crossentropy','bonangbarung9': 'binary_crossentropy', 'bonangbarung13': 'binary_crossentropy','bonangbarung17': 'binary_crossentropy','bonangbarung21': 'binary_crossentropy', 'bonangbarung25': 'binary_crossentropy','bonangbarung29': 'binary_crossentropy'},

              metrics={'bonangbarung1': 'accuracy','bonangbarung5': 'accuracy','bonangbarung9': 'accuracy','bonangbarung13': 'accuracy','bonangbarung17': 'accuracy','bonangbarung21': 'accuracy','bonangbarung25': 'accuracy','bonangbarung29': 'accuracy'})

model.summary()
```
